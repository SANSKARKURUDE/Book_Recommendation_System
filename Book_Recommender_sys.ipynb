# Popularity Based Recommender System

```python
ratings.merge(books,on = 'ISBN')
```

```python
ratings.merge(books,on = 'ISBN').shape
```

```python
book_ratings = ratings.merge(books,on = 'ISBN')
```

```python
book_ratings.head()
```

```python
no_of_ratings = book_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
no_of_ratings.rename(columns = {'Book-Rating' : 'No_Of_Ratings'},inplace = True)
no_of_ratings
```

```python
avg_ratings = book_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns = {'Book-Rating' : 'Avg_Ratings'},inplace = True)
avg_ratings
```

```python
Rated_books = no_of_ratings.merge(avg_ratings,on = 'Book-Title')
Rated_books.head()
```

```python
Popular_Books = Rated_books[Rated_books['No_Of_Ratings'] > 250].sort_values('Avg_Ratings',ascending = False)
```

```python
Popular_Books.merge(book_ratings,on = 'Book-Title').drop_duplicates('Book-Title')
```

# Collaborative Filtering Based Recommender System

```python
x = book_ratings.groupby('User-ID').count()['Book-Rating'] > 200
experienced_users = x[x].index
```

```python
book_ratings[book_ratings['User-ID'].isin(experienced_users)]
```

```python
ratings_by_exp_users = book_ratings[book_ratings['User-ID'].isin(experienced_users)]
```

```python
y = ratings_by_exp_users.groupby('Book-Title').count()['Book-Rating'] > 50
most_rated_books = y[y].index
```

```python
ratings_by_exp_users[ratings_by_exp_users['Book-Title'].isin(most_rated_books)]
```

```python
famous_books = ratings_by_exp_users[ratings_by_exp_users['Book-Title'].isin(most_rated_books)]
```

```python
final_df = famous_books.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
final_df.fillna(0,inplace = True)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_df)
```
